1. What is the distribution of age among heart failure patients in the dataset?

```python
sns.histplot(df.age,bins=28,color='r',edgecolor='yellow',kde=True)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

 2. How does the death rate vary with age?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='DEATH_EVENT', data=df, hue='DEATH_EVENT', palette='rainbow')
plt.title('Death Rate vs. Age')
plt.xlabel('Age')
plt.ylabel('Death Event (0: No, 1: Yes)')
plt.show()
```

3. What is the percentage of male and female patients in the dataset?

```python
gender_percentage = df['sex'].value_counts(normalize=True) * 100
print("Percentage of Male Patients:", gender_percentage[1])
print("Percentage of Female Patients:", gender_percentage[0])
```

```python
gender_counts = df['sex'].value_counts()

sizes = gender_counts.values
labels = ['Female', 'Male']
colors = ['skyblue', 'lightpink'] 
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=150)
plt.axis('equal') 
plt.title('Gender Distribution')
plt.show()
```

4. How does the platelet count vary among different age groups?

```python
plt.figure(figsize=(12, 6))
sns.displot(x='age', y='platelets', data=df, palette='Set3')
plt.title('Platelet Count Variation Among Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Platelet Count')
plt.show()
```

5. Is there a correlation between creatinine and sodium level in the blood?

```python
# Correlation between creatinine and sodium levels
plt.figure(figsize=(8, 6))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df, hue='sex', palette='Set1')
plt.title('Correlation between Creatinine and Sodium Levels')
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.legend(title='Gender', labels=['Female', 'Male'])
plt.show()
```

6. how does the prevalence of high blood pressure differ in male and female patients?

```python
# Calculate the percentage of patients with high blood pressure by gender
high_bp_by_gender = df.groupby('sex')['high_blood_pressure'].mean() * 100

# Plotting the prevalence of high blood pressure between male and female patients
plt.figure(figsize=(8, 6))
sns.barplot(x=high_bp_by_gender.index, y=high_bp_by_gender.values, color='skyblue')
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage with High Blood Pressure')
plt.xticks(ticks=[0, 1], labels=['Female', 'Male'])
plt.grid(True, axis='y')
plt.show()
```

Prevalence of High blood pressure by gender is greater in female

7. What is the relationship between smoking habits and occurrence of heart failure

```python
# Calculate the count of heart failure occurrence by smoking habits
heart_failure_by_smoking = df.groupby('smoking')['DEATH_EVENT'].value_counts().unstack().fillna(0)

# Plotting the relationship between smoking habits and heart failure occurrence
plt.figure(figsize=(8, 6))
sns.barplot(x=heart_failure_by_smoking.index, y=heart_failure_by_smoking[1], palette='husl')
plt.title('Heart Failure Occurrence by Smoking Habits')
plt.xlabel('Smoking')
plt.ylabel('Count of Heart Failure')
plt.xticks(ticks=[0, 1], labels=['Non-smoker', 'Smoker'])
plt.grid(True, axis='y')
plt.show()
```

8. Are there any noticeable patterns in the distribution of death events across different age groups

```python

# Plotting the distribution of death events across age groups
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['DEATH_EVENT'] == 1]['age'],color='red', alpha=0.5, label='Death Event')
sns.kdeplot(df[df['DEATH_EVENT'] == 0]['age'],color='blue', alpha=0.5, label='No Death Event')
plt.title('Distribution of Death Events Across Different Age Groups')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y')
plt.show()
```

1. Most of the death occur at the age of 60 to 80 years
2. No death event frequency is highest at 60 year old patients

9. Is there any significance difference between ejection fraction between patients with and without diabetes?

```python
ed = df[['ejection_fraction', 'diabetes']]

# Plotting the comparison of ejection fraction between patients with and without diabetes
plt.figure(figsize=(8, 6))
sns.boxplot(x='diabetes', y='ejection_fraction', data=ed, palette='Set2')
plt.title('Comparison of Ejection Fraction between Patients with and without Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Ejection Fraction')
plt.xticks(ticks=[0, 1], labels=['No Diabetes', 'Diabetes'])
plt.grid(True, axis='y')
plt.show()
```

1. Ejection Fraction is higher in non-diabetic patients than diabetic patients
2. IQR (Inter-Quartile range) is greater in diabetic patients
